```python
# https://github.com/clovaai/fewshot-font-generation/blob/main/docs/Dataset.md
# 1. prepare ttf files into same directory
# 2. split train and validation (train, val directory)
# 3. make available_characters to .txt file with DM-Font util(get_chars_from_ttf.py)
# 4. Prepare decomposition information(The files are provided)
#    Korean: data/kor/decomposition.json, data/kor/primals.json, data/kor/decomposition_DM.json(only for DM-Font)
# 5. Modify the data configuration file (cfgs/data/train/custom.yaml)

import os

ttf_dir = "/Users/oyutaeg/Desktop/ttf_changable"
file_list = [x for x in os.listdir(ttf_dir) if x.endswith(".ttf")]
```

```python
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import shutil
import os
```

```python
output_root_dir = "/home/ubuntu/ttf_dmfont"
x_train, x_valid, y_train, y_valid = train_test_split(file_list, file_list, test_size=0.2, shuffle=True, random_state=34)
```

```python
if not os.path.isdir(output_root_dir):
    os.mkdir(output_root_dir)
if not os.path.isdir(output_root_dir + os.path.sep +  "train"):
    os.mkdir(output_root_dir + os.path.sep +  "train")
if not os.path.isdir(output_root_dir + os.path.sep +  "valid"):
    os.mkdir(output_root_dir + os.path.sep +  "valid")
    
for ttf in  tqdm(x_train):
    shutil.copy(ttf_dir + ttf, output_root_dir + os.path.sep +  "train")
for ttf in  tqdm(x_valid):
    shutil.copy(ttf_dir + ttf, output_root_dir + os.path.sep + "valid")
    
```
